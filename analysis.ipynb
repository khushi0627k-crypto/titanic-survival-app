```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
df.groupby("Sex")["Survived"].mean()
```

```python
df.groupby("Sex")["Survived"].mean().plot(kind="bar")

plt.title("Survival Rate by Gender")
plt.ylabel("Survival Rate")
plt.show()
```

```python
df.groupby("Pclass")["Survived"].mean()
```

```python
df.groupby("Pclass")["Survived"].mean().plot(kind="bar")

plt.title("Survival Rate by Passenger Class")
plt.ylabel("Survival Rate")
plt.xlabel("Passenger Class")
plt.show()
```

```python
pd.crosstab(df["Pclass"], df["Sex"], values=df["Survived"], aggfunc="mean")
```

```python
pd.crosstab(df["Pclass"], df["Sex"], values=df["Survived"], aggfunc="mean").plot(kind="bar")

plt.title("Survival Rate by Class and Gender")
plt.ylabel("Survival Rate")
plt.show()
```

```python
df.isnull().sum()
```

```python
df = df.drop("Cabin", axis=1)
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
```

```python
df["Age"] = df["Age"].round()
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.groupby(pd.cut(df["Age"], bins=5))["Survived"].mean()
```

```python
df.groupby(pd.cut(df["Age"], bins=5))["Survived"].mean().plot(kind="bar")

plt.title("Survival Rate by Age Group")
plt.ylabel("Survival Rate")
plt.xlabel("Age Group")
plt.xticks(rotation=45)
plt.show()
```

```python
df.groupby(pd.cut(df["Age"], bins=5), observed=False)["Survived"].mean()
```

```python
df["AgeGroup"] = pd.cut(df["Age"], bins=[0,16,32,48,64,80])
df.head()
```
